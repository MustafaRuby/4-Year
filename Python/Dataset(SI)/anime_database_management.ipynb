#Search an anime using name as a subject

Motivation: when normal people generally search for an anime, they initially ask it's name, following the name, they may want to know more information about the anime, therefore a search based on the name is necessary.

```python
# Search anime using name

found_anime = None
anime_input = str(input("Insert the name of the anime: "))

print("\n")

for anime in data:
  if anime['title']['text'].lower() == anime_input.lower():
    found_anime = anime;
    break

if found_anime: # found_anime != null
  print(f"Name: {found_anime['title']['text']}; \ngenres: {found_anime['genres']}; \ndescription: {found_anime['description']}; \nstart time: {found_anime['start_date']}")
else:
  print("Anime not found (magnt u' tarall)")
```

#Search an anime using one genre as subject (D)

Motivation: most anime fans use genres as their base search material: some people might prefer fantasy type animes, while others may prefer action, and some may prefer even both.


```python
# Search anime using genre

found_anime = []  #The main list
anime_input = str(input("Insert the genre of the anime: "))   # The first genre to filter
print("\n")

for anime in data:  # Searches in data
  for genre in anime['genres']:
    if genre.lower() == anime_input.lower():
      found_anime.append(anime)
      break

if found_anime:
  for anime in found_anime:
    print(f"Name: {anime['title']['text']}; \ngenres: {anime['genres']}; \ndescription: {anime['description']}; \nstart time: {anime['start_date']}")
    print("\n-----------------------------------------------\n")
else:
  print("Anime not found (magnt u' tarall)")
```

#Search an anime using 2 genre as a subject

Motivation: most anime fans use genres as their base search material: some people might prefer fantasy type animes, while others may prefer action, and some may prefer even both.


```python
# Search anime using genre

found_anime_list = []  #The main list
found = False

anime_input = str(input("Insert the first genre of the anime: "))   # The first genre to filter
anime_input2 = str(input("Insert the second genre of the anime: "))   #The second genre to filter
print("\n")

for anime in data:  # Searches in data

  for genre in anime['genres']:   # Enters the genres of the anime in question

    if genre.lower() == anime_input.lower():  # Confronts the input with the genre

      for genre in anime['genres']:   # Has to check if the same anime has both the genres

        if genre.lower() == anime_input2.lower():   # Confronts

          found_anime_list.append(anime)   # Only then does it add
          found = True
          break

    elif genre.lower() == anime_input2.lower():

      for genre in anime['genres']:  # Has to check if the same anime has both the genres

          if genre.lower() == anime_input.lower():   # Confronts
            found_anime_list.append(anime)   # Only then does it add
            found = True
            break
    if found:
      break
  found = False

if found_anime_list:
  for anime in found_anime_list:
    print(f"Name: {anime['title']['text']}; \ngenres: {anime['genres']}; \ndescription: {anime['description']}; \nstart time: {anime['start_date']}")
    print("\n-----------------------------------------------\n")
else:
  print("Anime not found (magnt u' tarall)")
```

#Sorting of data alphabetically (D)

When you're going to a library or a place that sells the mangas, but also books, it's ordered alphabetically, therefore it's more akin to a standard.

# Genre graph

Creates a graph based on how many animes are there, that belong to a certain genre

```python
import matplotlib.pyplot as plt

genres_list = []  # The main list that will contain the genres

for anime in data:
  for genre in anime['genres']:
    if genre not in genres_list:
      genres_list.append(genre)

genres_numbers = {} # The dictionary, much needed for the graph | hashmap <String, Integer> String: genre, Integer: Number of anime

count = 0   # Will be of use to put a corresponding number of animes to the genres in the dictionary

# This following loop will generate the dictionary with the number of animes corresponding to the dictionary as value

for genre in genres_list:   # Goes into the main list

  for anime in data:  # Goes into data to check every anime

    for genre1 in anime['genres']:  # Goes into the genres of each anime

      if genre1 == genre:   # checks if the genre of the current anime is equal to the genre in the main list

        count += 1  # Increments the number if it's equal, at the end, this number will correspond to the number of animes of a specific genre

  genres_numbers[genre] = count # genres_number.get(genre)
  count = 0   # Resets the value of count (fundemental)

types = list(genres_numbers.keys())   # Corresponding to the genres
counts = list(genres_numbers.values())    # Corresponding to the number of animes corresponding

font1 = {'size':20}
plt.figure(figsize=(50,20))
plt.bar(types, counts, color = 'red')
plt.xlabel('Genre', fontdict = font1)
plt.ylabel('Number', fontdict = font1)
plt.title('Number of animes per genres', fontdict = font1)
plt.show()
```

#Dataset genres statistics



```python
print(f"Total number of animes in the dataset: {len(data)}")

print("\n-----------------------------------------------\n")

medium = {}   # Will to simplify the process of identifying the genre with maximum frequency in animes

for genre, count in genres_numbers.items():
  medium[genre] = (int(count)/len(genres_list))*100 # Adds the percentage corresponding to the genre as a value, while mantains the genre as a key, the order is irrelevant
  print(f"{medium[genre]} % of animes have the genre '{genre}'\n")

print("\n-----------------------------------------------\n")

max = ""    # Corresponding to the most frequent genre in animes of the list
max_count = 0   # Used to keep track of the most frequent genre

for genre, percentage in medium.items():
  if percentage > max_count:
    max_count = percentage
    max = str(genre)

print(f"Statistically the most frequent genre is is: {max}")
```
